## **NLTK**

```python
import nltk
nltk.download("punkt")
nltk.download('averaged_perceptron_tagger')
import requests

f=open("tur_wikipedia_2021_10K-sentences.txt","r",encoding="utf-8")
data=f.read()
```

corpus:https://wortschatz.uni-leipzig.de/en/download/Turkish


```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
stop_words = set(stopwords.words('turkish'))
```

```python
data[:100]
```

```python
from nltk import pos_tag
import string
```

```python
def noktalama_temiz(veri):

  metin = veri.translate(str.maketrans("", "", string.punctuation))
  return metin
```

```python
kelimeler=word_tokenize(noktalama_temiz(data))
kelimeler=[i for i in kelimeler if not i in stop_words]
pos_tag(kelimeler[:20])
```

```python
import zemberek
from zemberek.morphology import TurkishMorphology
from zemberek.tokenization import TurkishTokenizer
```

```python
tokenizer = TurkishTokenizer.DEFAULT #TurkishTokenizer.DEFAULT kullanılarak kelimeleri tokenlere ayırmak için tokenizer oluşturma
```

```python
morphology = TurkishMorphology.create_with_defaults() #Türkçe morfolojik analiz yapmak için nesne oluşturma
```

```python
sonuç = morphology.analyze("güzellik")
print(sonuç)
```

```python
tokenler = tokenizer.tokenize(data)
for token in tokenler:
    analiz = morphology.analyze(token.normalized)
    if analiz.analysis_results:
        print(analiz.analysis_results[0])
    else:
        print(f"analiz yok: {token}")
```
